```
# Importando Pandas. Biblioteca para análise de dados.
import pandas as pd 
# Importando as bibliotecas instaladas.
import requests 
import bs4
```

### Vamos aos trabalhos.

```
# Colocando a url que usaremos em uma constante.
URL = "https://suporte.boxloja.pro/article/90-faixa-de-ceps-do-brasil"

# Iniciando a requisição da url e já instanciando o BeautifulSoup para analisar o html da página.
web_page = bs4.BeautifulSoup(requests.get(URL).content, "lxml")

# Procuando a tag "table" no html da página.
table_cep = web_page.find(name="table")
table_cep
```

```
# Conferindo o titulo das colunas na tabela que estamos trabalhando.
colunas = table_cep.find_all(name="th")

t1 = colunas[0]
t2 = colunas[1]
t3 = colunas[2]

print(t1, t2, t3)
```

```
# Tratando nosso resultado. Colocando em uma lista os textos das tags html.
tags_text = [elem.text.strip("\n ") 
             for elem in table_cep.children 
             if type(elem) != bs4.NavigableString]

tags_text
```

```
# Mostrando nossa nova lista.
for i in tags_text:
  print(i)
```

```
# Aqui vemos que a nova lista contém apenas dois valores. 
len(tags_text)
```

```
# Titulos da tabela na primeira chave.
tags_text[0]
```

```
# Dados (Estados, Região, Faixa de Cep) na segunda chave
tags_text[1]
```

```
# Dividindo os titulos em uma outra lista. O critério de divisão será "\n".
titulo_tags_text = str(tags_text[0]).split('\n')
titulo_tags_text
```

```
# Dividindo os dados também em uma outra lista. Critério da divisão "\n\n\n".
info_tags_text = str(tags_text[1]).split('\n\n\n')
info_tags_text
```

```
# Verificando a quantidade de elementos da lista dos dados.
len(info_tags_text)
```

```
# Mostrando ordenadamente todas as informações que obtivemos.
for k, v in enumerate(info_tags_text):
  info_tags_text[k] = str(info_tags_text[k]).split('\n')
  print(k, v)
```

```
# Colocando em um DataFrame e plotanto as 10 primeiras informações.
df_teste = pd.DataFrame(info_tags_text)
df_teste.head(10)
# Vemos que a tabela não está em ordem. Iremos tratá-la.
```

Agora iremos inserir os "Estados" em seus devidos lugares. 

```
# Iremos pegar os índices exatamente onde cada um dos estados ínicia.
indice_estados = []
for k, v in enumerate(info_tags_text):
  if len(v) == 3: # Observado que onde temos o nome estado temos três valores. Então usaremos isto para nos guiar.
    indice_estados.append(k)
print(indice_estados)
```

```
# Obtendo o nome dos estados, colocando em uma nova lista, e excluindo da lista "info_tags_text".
lista_estados = []
for k, v in enumerate(info_tags_text):
  if len(v) == 3:
    lista_estados.append(info_tags_text[k][0])
    del(info_tags_text[k][0])

lista_estados
```

```
# Aqui estamos fazendo um ligação das lista dos índices(indice_estados) e dos estados(lista_estados). Colocaremos em um dicionário para melhor organização.
dicionario = {}
for i in range(0, len(lista_estados)):
  dicionario[indice_estados[i]] = lista_estados[i]

dicionario
```

```
# Esta parte é a mais importante da tratativa.
# Aqui iremos preencher os índices faltantes com seus devidos estados.
for i in range(1, len(info_tags_text)):
  if i not in dicionario.keys(): # Se o índice não estiver no dicionário, irá criar uma nova chave.
    dicionario[i] = dicionario[i-1] # O nome do estado que será preenchido será o nome do estado anterior.

# Veremos que adicionou tudo corretamente, porém, não está ordenado como gostaríamos.
print(dicionario)
```

```
# Um dicionário não pode ser ordenado através de um método, então, então, usaremos as artimanhas do python criando um novo dicionário.
dicionario_ordenado = {}
for i in range(0, len(dicionario.keys())):
  dicionario_ordenado[i] = dicionario[i]

# E vemos que agora está tudo como gostaríamos.
dicionario_ordenado
```

```
# Agora que tudo está em ordem, pegaremos apenas os nomes dos estados e colocaremos em uma nova lista.
estados = list(dicionario_ordenado.values())
estados
```

### Agora iremos colocar tudo em um DataFrame.

```
# Chamando o Pandas para ler nossa lista (info_tags_text) e transformar em um dataframe.
df = pd.DataFrame(info_tags_text)
```

```
# Mostrando as últimas 15 linahs da tabela.
df.tail(15)
```

```
# Vemos que o nome das nossas colunas está como "0"(localidades) e "1"(faixa de cep)
# Então colocaremos nomes nas respectivas colunas.
df.columns = ['Localidade', 'Faixa_Cep']

# Mostrando as 5 primeiras linhas.
df.head()
```

```
# Nossa tabela já está praticamente como queríamos, mas ainda falta os estados.
# Iremos inserir na posição "0" (primeira coluna da tabela), com o nome de "Estados", a lista dos estados.
df.insert(0, 'Estados', estados)
```
